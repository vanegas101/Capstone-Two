Change the data type of Release Date from object to datetime.

```python


def parse_dates(date):
    for fmt in ("%b %d, %Y", "%b %Y"):  
        try:
            return pd.to_datetime(date, format=fmt)
        except ValueError:
            continue
    return pd.NaT  


merged_df['Release Date'] = merged_df['Release Date'].apply(parse_dates)
```
